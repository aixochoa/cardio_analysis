# Conclusiones — Análisis Exploratorio de Enfermedad Cardíaca (Heart Disease UCI)

## Resumen del dataset

El análisis se realizó sobre 303 pacientes con 14 variables clínicas, sin valores faltantes. La muestra está razonablemente balanceada entre pacientes con enfermedad cardíaca confirmada (54%) y sin ella (46%), lo cual permite comparaciones confiables entre ambos grupos sin sesgo por tamaño de muestra.

## Hallazgos principales

**Los factores de riesgo clásicos no fueron los más predictivos.** Contrario a la intuición habitual, el colesterol sérico y la presión arterial en reposo mostraron correlaciones muy débiles con la presencia de enfermedad cardíaca confirmada (prácticamente nulas en el caso del colesterol). Esto sugiere que, en esta muestra, dichos factores funcionan más como predictores de riesgo a largo plazo que como marcadores de enfermedad ya presente.

**El comportamiento del corazón durante el esfuerzo fue mucho más informativo.** Las variables con mayor poder de discriminación fueron el tipo de dolor de pecho, la frecuencia cardíaca máxima alcanzada y la depresión del segmento ST inducida por ejercicio (oldpeak). Al combinar frecuencia cardíaca máxima y oldpeak en un mismo gráfico, se observó una separación visual clara entre pacientes con y sin enfermedad — algo que ninguna variable lograba por separado.

**El dolor de pecho "típico" fue, paradójicamente, el menos asociado a enfermedad confirmada.** La mayoría de los pacientes con angina típica (el síntoma de manual) no presentó enfermedad cardíaca confirmada, mientras que los dolores atípicos o no anginosos se asociaron fuertemente con diagnóstico positivo. Esto es consistente con la práctica clínica real, donde la presentación sintomática clásica no siempre coincide con la gravedad o presencia real de enfermedad coronaria.

**Edad y sexo mostraron patrones que requieren contexto.** Los pacientes sin enfermedad confirmada fueron, en promedio, mayores que los que sí la tenían — lo opuesto a lo esperado. De forma similar, las mujeres de la muestra presentaron una proporción de enfermedad mucho mayor que los hombres, a pesar de ser un grupo más pequeño. Ambos hallazgos probablemente reflejan un sesgo de selección propio del dataset (pacientes ya derivados a estudio cardiológico por sospecha clínica), más que una relación causal directa con la edad o el sexo en la población general.

## Limitaciones

Este es un análisis exploratorio descriptivo, no causal. El dataset proviene de una población clínica ya seleccionada (no población general), lo cual limita la generalización de los hallazgos. Las correlaciones observadas no implican causalidad y deben interpretarse en ese contexto.
